# Access Modifiers

## Modifiers in Python
*Using no `underscores` to identify public Attribute or Method in Python* <br />
*Using single underscore `_` to identify protected Attribute or Method in Python* <br />
*Using double underscore `__` to identify private Attribute or Method in Python* <br />

## Protected Modifier

```python
class Human:
    def __init__(self,value=None):
        self._protectedAttribute = value
    def _protectedMethod(self):
        print(f'Protected Attribute is {self._protectedAttribute}')
```

```python
human = Human('Ahmed')
```

```python
human._protectedMethod()
```

## Access Parent's Private Attributes / Methods from Child Class

In sub-class (Child) you have to write `_<Parent>` replaced with class name followed by `__<Private>` replaced with attribute or method name

#### Example:

```python
class SupClass:
    def __privateWelcome(self):
        print('Hello from private method in Super-class')
```

```python
class SubClass(SupClass):
    def hello(self):
        self._SupClass__privateWelcome()
```

```python
sup = SubClass()
sup.__privateWelcome()
```

```python
sub = SubClass()
sub.hello()
```
